Develop a classification model for Iris dataset using  Pseudo  inversemethod  and evaluate the  performance. 

```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
import matplotlib.pyplot as plt
```

```python
iris.feature_names 
```

```python
iris.target_names
```

```python
#creating a dataframe with columns as feature names
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
#setting the labels of the dataframe

df['target']=iris.target
df.head
```

```python
#Splitting the data as input and the target
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

ONE HOT ENCODING

```python
#one hot encoding the target names as 1 or 0.
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder()
#enc.fit(y)
y=enc.fit_transform(Y.values.reshape(-1,1)).toarray()
  
#Merge with main
#y=df.join(enc_df)
  
print(y)


```

TRAIN-TEST SPLIT

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
print(X_train.shape)
```

```python
#finding the weight matrix
w = np.matmul(np.linalg.pinv(X_train),y_train)
```

```python
w
```

```python
y_pred = np.matmul(X_test,w)
y_pred
```

```python
#find out the highest value for each row 
maxVal = np.asarray(y_pred.max(axis=1))
maxVal
```

```python
# for one hot encoding of the predicted value ,i.e. highest value in each row will be 1 and other will be 0.
for i in range(len(maxVal)):
    for j in range(3):
        
        if y_pred.iloc[i,j] == maxVal[i]:
            y_pred.iloc[i,j] = 1
        else :
            y_pred.iloc[i,j] = 0
            
y_pred.head()
```

```python
#Accuracy measures
from sklearn.metrics import accuracy_score,classification_report,precision_score,recall_score,f1_score

print('Accuracy Score : ', accuracy_score(y_test,y_pred))
print('Precision Score : ', precision_score(y_test,y_pred, average = 'weighted'))
print('Recall Score : ', recall_score(y_test,y_pred,average = 'weighted'))
print('F1 Score : ', f1_score(y_test,y_pred,average = 'weighted'))
print('\n')
print(classification_report(y_test,y_pred))
```
